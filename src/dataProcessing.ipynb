```python
import pandas as pd
import numpy as np
```

```python
unCleanData = pd.read_csv('../data/SkyrimWeapons.csv')
unCleanData.head()
```

Remove speed col as most of it is NaN

```python
unCleanData = unCleanData.iloc[:,:-1]   # select everything except speed col
unCleanData
```

Remove silverware, arrows and bolts

```python
unCleanData = unCleanData.loc[(unCleanData['Type'] != 'Bolt') & (unCleanData['Type'] != 'Arrow') & (unCleanData['Type'] != 'Silverware')]
```

```python
unCleanData['Upgrade'] = unCleanData['Upgrade'].replace(to_replace=np.NaN, value='No Upgrade')
unCleanData
```

Drop perk because most of the rows arent filled

```python
unCleanData.pop('Perk')
```

```python
cleanedData = unCleanData
```

```python
pd.DataFrame.to_csv(cleanedData, '../data/CleanSkyrimData.csv')
```
